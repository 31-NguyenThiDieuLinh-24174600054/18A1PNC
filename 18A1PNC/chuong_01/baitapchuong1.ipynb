```python
class hinhchunhat:
    # khởi tạo hàm (contruction) gồm 2 thuộc tính : chiều dài , chiều rộng
    def __init__(self, dai=0, rong=0):
        self.dai=dai
        self.rong=rong
    # phương thức nhập dữ liệu
    def nhap(self):
        self.dai=float(input("nhập chiều dài: "))
        self.rong=float(input("nhap chiều rộng: "))
    # tính chu vi hình chữ nhật
    def chu_vi(self):
        p=(self.dai+self.rong)*2
        print("chu vi là",p)
    # tính diện tích hình chữ nhật
    def dien_tich(self):
        s=self.dai*self.rong
        print("diện tích là", s)
hcn=hinhchunhat()
hcn.nhap()
hcn.chu_vi()
hcn.dien_tich()
```

```python
class thisinh:
    # khởi tạo hàm với họ tên và điểm 3 môn
    def __init__(self, sobd, hoten, toan, ly, hoa):
        self.sobd=sobd
        self.hoten=hoten
        self.toan=toan
        self.ly=ly
        self.hoa=hoa
    # tính tổng điểm 3 môn
    def tong_diem(self):
        return self.toan+self.ly+self.hoa
    def thong_tin(self):
        print(f"{self.sobd<10}{self.hoten<30}{self.toan<8}{self.toan<8}{self.ly<8}{self.hoa<8}{self.tong_diem<8}")
# nhập danh sách thí sinh
danh_sach=[]
n=int(input("nhập số lượng thí sinh: "))
for i in range(n):
    sobd=input("số báo danh: ")
    hoten=input("nhập họ và tên: ")
    toan=float(input("nhập điểm toán: "))
    ly=float(input("nhập điểm lý: "))
    hoa=float(input("nhập điểm hoa: "))
    danh_sach.append(thisinh(sobd, hoten, toan, ly, hoa))
# sắp xếp thí sinh từ cao đến thápa
danh_sach.sort(key=lambda thisinh: thisinh.tong_diem(),reverse=True)
#in danh sách và xác định trúng tuyển
print(f"\n {'SBD':<10}{'Họ tên':<30}{'Toán':<8}{'Lý':<8}{'Hóa':<8}{'Tổng':<8}")
for thisinh in danh_sach:
    print(f"{thisinh.sobd:<10}{thisinh.hoten:<30}{thisinh.toan:<8}{thisinh.ly:<8}{thisinh.hoa:<8}{thisinh.tong_diem():<8}")
# điểm chuẩn đầu vào là 20
print("\n Danh sách thí sinh trúng tuyển là: ")
for thisinh in danh_sach:
    if thisinh.tong_diem()>=20:
        print(f"→ {thisinh.hoten} (Tổng điểm: {thisinh.tong_diem()})")
```

```python
class Stack:
    # Hàm tạo 
    def __init__(self, capacity):
        self.capacity = capacity             # sức chứa tối đa
        self.items = [0.0] * capacity        # mảng float cấp phát động
        self.top = -1                        # chỉ số phần tử trên cùng
        print("Ngăn xếp đã được tạo thành công!")
    # Hàm hủy 
    def __del__(self):
        print("Ngăn xếp đã bị hủy và xóa khỏi bộ nhớ!")
    # Kiểm tra ngăn xếp có rỗng không
    def isEmpty(self):
        return self.top == -1
    # Kiểm tra ngăn xếp có đầy không
    def isFull(self):
        return self.top == self.capacity - 1
    # Đưa một phần tử vào ngăn xếp
    def push(self, value):
        if self.isFull():
            print("Ngăn xếp đã đầy, không thể thêm phần tử!")
        else:
            self.top += 1
            self.items[self.top] = float(value)
            print(f"Đã thêm {value} vào ngăn xếp.")
    # Lấy một phần tử ra khỏi đỉnh ngăn xếp
    def pop(self):
        if self.isEmpty():
            print("Ngăn xếp đang rỗng, không thể lấy phần tử!")
            return None
        else:
            value = self.items[self.top]
            self.top -= 1
            print(f"Đã lấy {value} ra khỏi ngăn xếp.")
            return value
    # In ngăn xếp
    def display(self):
        if self.isEmpty():
            print("Ngăn xếp trống.")
        else:
            print("Các phần tử trong ngăn xếp:", self.items[:self.top + 1])
# sử dụng
if __name__ == "__main__":
    s = Stack(3)          # tạo ngăn xếp có 3 phần tử tối đa
    s.push(1.5)
    s.push(2.2)
    s.push(3.8)
    s.push(4.0)          # sẽ báo đầy
    s.display()
    s.pop()
    s.display()
    s.pop()
    s.pop()
    s.pop()               # sẽ báo rỗng
```

```python
class Stack:
    # Hàm khởi tạo - tạo ngăn xếp rỗng
    def __init__(self):
        self.items = []
        print("Stack đã được tạo thành công!")
    # Hàm hủy
    def __del__(self):
        print("Stack đã bị hủy và xóa khỏi bộ nhớ!")
    # Hàm push - thêm phần tử vào ngăn xếp
    def push(self, item):
        self.items.append(item)
        print(f"Đã thêm phần tử: {item}")
    # Hàm pop - xóa phần tử ở đỉnh ngăn xếp
    def pop(self):
        if not self.is_empty():
            removed = self.items.pop()
            print(f"Đã lấy ra phần tử: {removed}")
            return removed
        else:
            print("Ngăn xếp rỗng, không thể pop!")
            return None
    # Hàm kiểm tra ngăn xếp rỗng
    def is_empty(self):
        return len(self.items) == 0
    #  Hàm print - in nội dung ngăn xếp
    def print(self):
        if self.is_empty():
            print("Ngăn xếp hiện đang rỗng.")
        else:
            print("Nội dung ngăn xếp (từ dưới lên trên):")
            for i, item in enumerate(self.items):
                print(f"{i+1}: {item}")
#sử dụng
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.print()    # In ra toàn bộ nội dung ngăn xếp
s.pop()
s.print()

```

```python
class Date:
    # Hàm khởi tạo date
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year
    # Hàm display: in thông tin về day, month, year
    def display(self):
        print(f"{self.day:02}/{self.month:02}/{self.year}")
    # Hàm kiểm tra năm nhuận
    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)
    # Hàm cho biết số ngày trong tháng
    def day_in_month(self):
        if self.month in [1,3,5,7,8,10,12]:
            return 31
        elif self.month in [4,6,9,11]:
            return 30
        elif self.month == 2:
            if self.is_leap_year():
                return 29
            else:
                return 28
        else:
            return 0  # Tháng không hợp lệ
    # Hàm next: tính ngày tiếp theo
    def next_day(self):
        self.day += 1  # tăng thêm 1 ngày
        # nếu vượt quá số ngày trong tháng thì reset về 1 và tăng thêm 1 tháng
        if self.day > self.day_in_month():
            self.day = 1
            self.month += 1
        # nếu tháng vượt quá 12 thì qua năm mới
        if self.month > 12:
            self.month = 1
            self.year += 1
# sử dụng
d = Date(6, 4, 2006)
d.display()        
d.next_day()
d.display()        
d.next_day()
d.display()   
```

```python
import math
#  lớp cha
class Tamgiac:
    def __init__(self, a, b, c):
        self.a=a
        self.b=b
        self.c=c
    # tính chu vi tam giá
    def chu_vi(self):
        return self.a+self.b+self.c
    #tính diện tích tam giác theo công thức heron
    def dien_tich(self):
        p=self.chu_vi()/2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
# lớp con 1
class Tamgiacvuong:
    def __init__(self, a, b, c):
        self.a=a
        self.b=b
        self.c=c
    #tính chu vi 
    def chu_vi(self):
        return self.a+self.b+self.c
    #tính diện tích
    def dien_tich(self):
        return (self.a*self.b)/2
#lớp con 3
class Tamgiaccan(Tamgiac):
    def __init__(self, a, b, h):
        self.a=a
        self.b=b
        self.h=h
    #tính chu vi
    def chu_vi(self):
        return (2*self.a)+self.b
    #tính diện tích
    def dien_tich(self):
        return (self.a*self.h)/2
    
#lớp con 4
class Tamgiacdeu(Tamgiaccan):
    def __init__(self, a):
        self.a=a
    #tính chu vi
    def chu_vi(self):
        return 3*self.a
    #tính diện tichs
    def dien_tich(self):
        return (math.sqrt(3)/4)*self.a**2
# sử dụng
# Tam giác thường
tg = Tamgiac(3, 4, 5)
print("Tam giác thường:")
print("Chu vi:", tg.chu_vi())
print("Diện tích:", tg.dien_tich())

# Tam giác vuông
tgv = Tamgiacvuong(3, 4, 5)
print("\nTam giác vuông:")
print("Chu vi:", tgv.chu_vi())
print("Diện tích:", tgv.dien_tich())

# Tam giác cân
tgc = Tamgiaccan(5, 6, 4)
print("\nTam giác cân:")
print("Chu vi:", tgc.chu_vi())
print("Diện tích:", tgc.dien_tich())

# Tam giác đều
tgd = Tamgiacdeu(6)
print("\nTam giác đều:")
print("Chu vi:", tgd.chu_vi())
print("Diện tích:", tgd.dien_tich())
```
