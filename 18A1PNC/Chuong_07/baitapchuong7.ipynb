```python
import tkinter as tk

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Student Form")
        self.root.geometry('350x80')

        self.label1 = tk.Label(root, text="Student Name:")
        self.label1.grid(column=0, row=0)

        self.txt1 = tk.Entry(root, width=15)
        self.txt1.grid(column=1, row=0)
        self.txt1.focus()

        self.label2 = tk.Label(root, text="Student ID:")
        self.label2.grid(column=0, row=1)

        self.txt2 = tk.Entry(root, width=15)
        self.txt2.grid(column=1, row=1)

        self.label3 = tk.Label(root, text="Password:")
        self.label3.grid(column=0, row=2)
        self.txt3 = tk.Entry(root, width=15, show="*") 
        self.txt3.grid(column=1, row=2)

        self.submit_btn = tk.Button(root, text="Submit", command=self.check_entries)
        self.submit_btn.grid(column=0, row=3, columnspan=2, pady=5)

    def check_entries(self):
        name = self.txt1.get()
        id_ = self.txt2.get()
        password = self.txt3.get()
       
        if name and id_ and password:
            print(f"Submitted: Name={name}, ID={id_}, Password={password}")
           
        else:
            print("Please fill all fields.")

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import requests
from io import BytesIO

class ImageViewer:
    def __init__(self, root):
        self.root = root
        self.root.title("Image Viewer")
        self.root.geometry("600x500")

        # Label để hiển thị ảnh
        self.image_label = tk.Label(root)
        self.image_label.pack(pady=10)

        # Frame cho các nút
        button_frame = tk.Frame(root)
        button_frame.pack(pady=5)

        # Nút chọn file
        self.load_file_btn = tk.Button(button_frame, text="Load from File", command=self.load_from_file)
        self.load_file_btn.pack(side=tk.LEFT, padx=5)

        # Entry và nút cho URL
        self.url_entry = tk.Entry(button_frame, width=40)
        self.url_entry.pack(side=tk.LEFT, padx=5)
        self.load_url_btn = tk.Button(button_frame, text="Load from URL", command=self.load_from_url)
        self.load_url_btn.pack(side=tk.LEFT, padx=5)

        # Tự động tải 'Otto.png' nếu tồn tại
        self.load_default_image()

    def load_default_image(self):
        # Thử tải 'Otto.png' từ thư mục hiện tại
        try:
            img = Image.open('Otto.png')
            self.display_image(img)
        except FileNotFoundError:
            messagebox.showinfo("Info", "File 'Otto.png' not found. Please load an image manually.")
        except Exception as e:
            messagebox.showerror("Error", f"Failed to load 'Otto.png': {str(e)}")

    def load_from_file(self):
        # Mở dialog chọn file, hỗ trợ .jpg, .png, .gif
        file_path = filedialog.askopenfilename(
            filetypes=[("Image files", "*.jpg *.jpeg *.png *.gif")]
        )
        if file_path:
            try:
                img = Image.open(file_path)
                self.display_image(img)
            except Exception as e:
                messagebox.showerror("Error", f"Failed to load image: {str(e)}")

    def load_from_url(self):
        url = self.url_entry.get().strip()
        if not url:
            messagebox.showwarning("Warning", "Please enter a URL.")
            return
        try:
            response = requests.get(url)
            response.raise_for_status()
            img = Image.open(BytesIO(response.content))
            self.display_image(img)
        except requests.exceptions.RequestException as e:
            messagebox.showerror("Error", f"Failed to load image from URL: {str(e)}")
        except Exception as e:
            messagebox.showerror("Error", f"Failed to process image: {str(e)}")

    def display_image(self, img):
        # Resize ảnh nếu quá lớn để vừa cửa sổ (tùy chọn)
        img.thumbnail((500, 400), Image.Resampling.LANCZOS)
        img_tk = ImageTk.PhotoImage(img)
        self.image_label.config(image=img_tk)
        self.image_label.image = img_tk  # Giữ tham chiếu để tránh garbage collection

if __name__ == "__main__":
    root = tk.Tk()
    app = ImageViewer(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class WordReverser:
    def __init__(self, root):
        self.root = root
        self.root.title("Word Validate")
        self.root.geometry("400x200")

        # Label hướng dẫn
        self.label = tk.Label(root, text="Enter a word:")
        self.label.pack(pady=10)

        # Entry để nhập từ
        self.entry = tk.Entry(root, width=40)
        self.entry.pack(pady=5)
        self.entry.focus()

        # Nút để đảo ngược
        self.validate_btn = tk.Button(root, text="Validate", command=self.validate_word)
        self.validate_btn.pack(pady=10)

        # Label để hiển thị kết quả
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=10)

    def validate_word(self):
        word = self.entry.get().strip()
        if not word:
            messagebox.showwarning("Warning", "Please enter a word.")
            return
        
        # Đảo ngược mà không dùng built-in (tương tự bài trước)
        validate_word = ""
        for i in range(len(word) - 1, -1, -1):
            validate_word += word[i]
        
        # Hiển thị kết quả
        self.result_label.config(text=f"Validate: {validate_word}")

if __name__ == "__main__":
    root = tk.Tk()
    app = WordReverser(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import math

class GCDLCMCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("GCD and LCM Calculator")
        self.root.geometry("450x300")

        # Label hướng dẫn
        self.label = tk.Label(root, text="Enter two integers A and B:")
        self.label.pack(pady=10)

        # Frame cho các entry
        entry_frame = tk.Frame(root)
        entry_frame.pack(pady=5)

        # Entry cho A
        tk.Label(entry_frame, text="Entervalue of of m:").grid(row=0, column=0, padx=5)
        self.entry_a = tk.Entry(entry_frame, width=15)
        self.entry_a.grid(row=0, column=1, padx=5)
        self.entry_a.focus()

        # Entry cho B
        tk.Label(entry_frame, text="Entervalue of of n:").grid(row=1, column=0, padx=5)
        self.entry_b = tk.Entry(entry_frame, width=15)
        self.entry_b.grid(row=1, column=1, padx=5)

        # Nút để tính toán
        self.calculate_btn = tk.Button(root, text="Calculate GCD and LCM", command=self.calculate)
        self.calculate_btn.pack(pady=15)

        # Label để hiển thị kết quả
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=10)

    def calculate(self):
        try:
            a = int(self.entry_a.get().strip())
            b = int(self.entry_b.get().strip())
            
            if a == 0 and b == 0:
                raise ValueError("At least one number must be non-zero.")
            
            # Tính GCD và LCM
            gcd_value = math.gcd(a, b)
            lcm_value = abs(a * b) // gcd_value if gcd_value != 0 else 0
            
            # Hiển thị kết quả
            self.result_label.config(text=f"GCD of {a} and {b}: {gcd_value}\nLCM of {a} and {b}: {lcm_value}")
        except ValueError:
            messagebox.showerror("Error", "Please enter valid integers for A and B.")
            self.result_label.config(text="")

if __name__ == "__main__":
    root = tk.Tk()
    app = GCDLCMCalculator(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class GreetingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Greeting App")
        self.root.geometry("400x250")

        # Label hướng dẫn
        self.label = tk.Label(root, text="Enter your name and age:")
        self.label.pack(pady=10)

        # Frame cho các entry
        entry_frame = tk.Frame(root)
        entry_frame.pack(pady=5)

        # Entry cho tên
        tk.Label(entry_frame, text="Name:").grid(row=0, column=0, padx=5)
        self.name_entry = tk.Entry(entry_frame, width=20)
        self.name_entry.grid(row=0, column=1, padx=5)
        self.name_entry.focus()

        # Entry cho tuổi
        tk.Label(entry_frame, text="Age:").grid(row=1, column=0, padx=5)
        self.age_entry = tk.Entry(entry_frame, width=20)
        self.age_entry.grid(row=1, column=1, padx=5)

        # Nút để submit
        self.submit_btn = tk.Button(root, text="Submit", command=self.generate_greeting)
        self.submit_btn.pack(pady=15)

        # Label để hiển thị kết quả
        self.result_label = tk.Label(root, text="", font=("Arial", 12), wraplength=350)
        self.result_label.pack(pady=10)

    def generate_greeting(self):
        try:
            name = self.name_entry.get().strip()
            age = int(self.age_entry.get().strip())
            
            if not name:
                raise ValueError("Name cannot be empty.")
            if age < 0:
                raise ValueError("Age must be a non-negative integer.")
            
            # Tạo thông báo dựa trên tuổi
            if age >= 18:
                message = f"Xin chào {name}. Bạn đã trưởng thành! {age}"
            else:
                message = f"Xin chào {name}. Bạn còn nhỏ tuổi!"
            
            # Hiển thị kết quả
            self.result_label.config(text=message)
        except ValueError as e:
            messagebox.showerror("Error", str(e))
            self.result_label.config(text="")

if __name__ == "__main__":
    root = tk.Tk()
    app = GreetingApp(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import os

def tinh_am_lich():
    try:
        nam_duong = int(entry_nam.get())
        if nam_duong < 0: raise ValueError
        
        can = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
        chi = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]
        
        # Công thức tính Can Chi dựa trên số dư cho 10 và 12
        ten_can = can[nam_duong % 10]
        ten_chi = chi[nam_duong % 12]
        nam_am = f"{ten_can} {ten_chi}"
        
        # Hiển thị thông báo kết quả
        messagebox.showinfo("Kết quả", f"Năm Dương lịch {nam_duong} là năm: {nam_am}")
        
        # Hiển thị hình ảnh con giáp (Chi)
        hien_thi_con_giap(ten_chi)
        
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập năm là một số nguyên dương!")

def hien_thi_con_giap(ten_chi):
    # Giả sử bạn có các file ảnh đặt tên là Ty.png, Suu.png... trong thư mục con 'images'
    # Nếu không có ảnh, phần này sẽ báo lỗi nhưng không làm sập chương trình
    try:
        path = f"images/{ten_chi.lower()}.png" 
        img = Image.open(path)
        img = img.resize((150, 150))
        photo = ImageTk.PhotoImage(img)
        lbl_img.config(image=photo)
        lbl_img.image = photo
    except:
        lbl_img.config(text=f"(Chưa có ảnh cho tuổi {ten_chi})", image='')

# Thiết lập giao diện
root = tk.Tk()
root.title("Bài 7.11: Đổi năm Dương lịch sang Âm lịch")
root.geometry("400x400")

tk.Label(root, text="Nhập năm Dương lịch:", font=("Arial", 12)).pack(pady=10)
entry_nam = tk.Entry(root, font=("Arial", 12), justify='center')
entry_nam.pack(pady=5)

btn_convert = tk.Button(root, text="Chuyển đổi", command=tinh_am_lich, bg="orange", fg="white", font=("Arial", 10, "bold"))
btn_convert.pack(pady=20)

lbl_img = tk.Label(root)
lbl_img.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk

def tinh_bmi():
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        bmi = weight / (height ** 2)
        
        lbl_bmi_val.config(text=f"BMI: {bmi:.2f}")
        
        # Phân loại kết luận
        if bmi < 18.5: res = "Gầy"
        elif 18.5 <= bmi < 24.9: res = "Bình thường"
        elif 25 <= bmi < 29.9: res = "Tiền béo phì"
        else: res = "Béo phì"
        
        lbl_result.config(text=f"Kết luận: {res}")
    except ValueError:
        lbl_result.config(text="Lỗi: Nhập sai số liệu!")

root = tk.Tk()
root.title("BMI Calculator")
root.geometry("300x250")

tk.Label(root, text="Cân nặng (kg):").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (m):").pack(pady=5)
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=tinh_bmi).pack(pady=10)

lbl_bmi_val = tk.Label(root, text="BMI: --")
lbl_bmi_val.pack()

lbl_result = tk.Label(root, text="Kết luận: --", font=("Arial", 10, "bold"))
lbl_result.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk

def tinh_tien():
    try:
        km = float(entry_km.get())
        wait_time = int(entry_wait.get())
        car_type = var_car.get()
        
        # Tính tiền chờ: 5p đầu free, từ phút thứ 6 là 750đ/phút
        tien_cho = max(0, wait_time - 5) * 750
        
        if car_type == 4: # Xe 4 chỗ
            if km <= 0.8: total = 12000
            elif km <= 30: total = 12000 + (km - 0.8) * 15300
            else: total = 12000 + (30 - 0.8) * 15300 + (km - 30) * 12100
        else: # Xe 7 chỗ
            if km <= 0.8: total = 12000
            elif km <= 30: total = 12000 + (km - 0.8) * 16100
            else: total = 12000 + (30 - 0.8) * 16100 + (km - 30) * 13800
            
        lbl_total.config(text=f"Tổng tiền: {int(total + tien_cho):,} đồng")
    except ValueError:
        lbl_total.config(text="Vui lòng nhập số hợp lệ!")

root = tk.Tk()
root.title("Tính cước Taxi")

tk.Label(root, text="Nhập số km:").pack()
entry_km = tk.Entry(root)
entry_km.pack()

tk.Label(root, text="Thời gian chờ (phút):").pack()
entry_wait = tk.Entry(root)
entry_wait.pack()

var_car = tk.IntVar(value=4)
tk.Radiobutton(root, text="Xe 4 chỗ", variable=var_car, value=4).pack()
tk.Radiobutton(root, text="Xe 7 chỗ", variable=var_car, value=7).pack()

tk.Button(root, text="Tính tiền", command=tinh_tien).pack(pady=10)
lbl_total = tk.Label(root, text="Tổng tiền: 0 đồng", font=("Arial", 11, "bold"))
lbl_total.pack()

root.mainloop()
```
