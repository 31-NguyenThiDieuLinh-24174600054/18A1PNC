```python
import numpy as np

# 1. Đọc dữ liệu vào numpy array 2D
np_baseball = np.loadtxt("baseball_2D.txt", dtype=int)
print("Dữ liệu np_baseball:\n", np_baseball)
print("Kiểu dữ liệu:", type(np_baseball))
print("Kích thước:", np_baseball.shape)  
# 2. 
print("Giá trị dòng thứ 50:", np_baseball[49])

# 3. 
np_weight = np_baseball[:, 1]
print("Mảng cân nặng:", np_weight)

# 4. 
print("Chiều cao vận động viên thứ 124:", np_baseball[123, 0], "inch")

# 5. 
print("Chiều cao trung bình:", np.mean(np_baseball[:, 0]), "inch")
print("Cân nặng trung bình:", np.mean(np_baseball[:, 1]), "pound")

# 6.
corr = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("Hệ số tương quan:", corr)

if corr > 0:
    print("→ Chiều cao và cân nặng có tương quan **thuận**: cao hơn → nặng hơn.")
elif corr < 0:
    print("→ Chiều cao và cân nặng có tương quan **nghịch**.")
else:
    print("→ Không có mối liên hệ rõ ràng.")
```

```python
import numpy as np

teams = np.array(['Germany', 'Greece', 'England', 'Italy', 'Russia', 'Spain', 'Portugal', 'France'])
goals = np.array([12, 4, 7, 9, 6, 10, 5, 3])
yellow_cards = np.array([5, 7, 6, 4, 3, 2, 5, 6])
red_cards = np.array([1, 0, 2, 1, 0, 1, 0, 1])
shooting_accuracy = np.array([52.5, 43.0, 49.5, 50.2, 45.3, 55.1, 48.0, 46.2])

euro12 = np.zeros(len(teams), dtype=[('Team', 'U20'), 
                                     ('Goals', 'i4'), 
                                     ('Yellow Cards', 'i4'), 
                                     ('Red Cards', 'i4'),
                                     ('Shooting Accuracy', 'f4')])

euro12['Team'] = teams
euro12['Goals'] = goals
euro12['Yellow Cards'] = yellow_cards
euro12['Red Cards'] = red_cards
euro12['Shooting Accuracy'] = shooting_accuracy

print("Type:", type(euro12))
print("Shape:", euro12.shape)
print("Columns:", euro12.dtype.names)


#1
print("\nGoals column:")
print(euro12['Goals'])
#2
print("\nSố đội tham gia:", len(np.unique(euro12['Team'])))

#3
print("\nThông tin DataFrame:")
for name in euro12.dtype.names:
    print(f"{name}: {euro12[name].dtype}, shape: {euro12[name].shape}")

#4
discipline = np.column_stack((euro12['Team'], euro12['Yellow Cards'], euro12['Red Cards']))

# 5
red_cards = discipline[:,2].astype(int)
yellow_cards = discipline[:,1].astype(int)
sorted_indices = np.lexsort((yellow_cards, red_cards))[::-1]  
discipline_sorted = discipline[sorted_indices]
print("\nDiscipline sorted:")
print(discipline_sorted)

#6a
avg_yellow = np.mean(euro12['Yellow Cards'])
print("\nTrung bình Yellow Cards:", avg_yellow)

#6b
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]['Team']
print("\nĐội ghi hơn 6 bàn thắng:")
print(teams_more_than_6_goals)

#7
teams_start_G = [team for team in euro12['Team'] if team.startswith('G')]
print("\nĐội bắt đầu bằng 'G':")
print(teams_start_G)

#8
print("\n7 cột đầu:")
first_7_cols = [euro12[name] for name in euro12.dtype.names[:7]]
print(np.column_stack(first_7_cols))

#9
print("\nTất cả cột trừ 3 cột cuối:")
cols_except_last3 = [euro12[name] for name in euro12.dtype.names[:-3]]
print(np.column_stack(cols_except_last3))

#10
teams_mask = np.isin(euro12['Team'], ['England', 'Italy', 'Russia'])
selected_data = euro12[teams_mask]
print("\nCác cột 'Team', 'Goals', 'Shooting Accuracy' từ 3 đội:")
for row in selected_data:
    print(row['Team'], row['Goals'], row['Shooting Accuracy'])

#11
print("\nChỉ 'Team' và 'Shooting Accuracy' từ 3 đội:")
for row in selected_data:
    print(row['Team'], row['Shooting Accuracy'])

```

```python
import numpy as np

countries = np.array(['USA','France','Germany','India','Nigeria','Brazil','China','Australia','Russia','Italy'])
beer_servings = np.array([249, 127, 346, 12, 25, 245, 45, 311, 200, 175])
spirit_servings = np.array([158, 120, 70, 45, 40, 130, 100, 150, 210, 90])
wine_servings = np.array([50, 220, 175, 10, 20, 70, 15, 180, 90, 200])
total_litres_of_pure_alcohol = np.array([8.5, 12.3, 10.5, 3.2, 2.8, 7.0, 4.1, 11.5, 9.0, 10.8])
continent = np.array(['North America','Europe','Europe','Asia','Africa','South America','Asia','Oceania','Europe','Europe'])


drink = np.zeros(len(countries), dtype=[('country','U20'),
                                        ('beer_servings','i4'),
                                        ('spirit_servings','i4'),
                                        ('wine_servings','i4'),
                                        ('total_litres_of_pure_alcohol','f4'),
                                        ('continent','U20')])

drink['country'] = countries
drink['beer_servings'] = beer_servings
drink['spirit_servings'] = spirit_servings
drink['wine_servings'] = wine_servings
drink['total_litres_of_pure_alcohol'] = total_litres_of_pure_alcohol
drink['continent'] = continent

#1
print("Type:", type(drink))
print("Shape:", drink.shape)
print("Columns:", drink.dtype.names)

print("\n5 dòng đầu:")
print(drink[:5])

print("\n5 dòng cuối:")
print(drink[-5:])

#2
continents = np.unique(drink['continent'])
avg_beer_by_continent = {c: drink['beer_servings'][drink['continent']==c].mean() for c in continents}
print("\nBia tiêu thụ trung bình theo châu lục:")
for k,v in avg_beer_by_continent.items():
    print(k, ":", v)

#3
print("\nThống kê rượu vang theo châu lục:")
for c in continents:
    wine = drink['wine_servings'][drink['continent']==c]
    print(f"{c}: mean={wine.mean():.2f}, median={np.median(wine)}, min={wine.min()}, max={wine.max()}, std={wine.std():.2f}")

# =========================
# 4. Trung bình beer + wine theo châu lục
# =========================
print("\nTrung bình bia và rượu vang theo châu lục:")
for c in continents:
    mask = drink['continent']==c
    mean_beer = drink['beer_servings'][mask].mean()
    mean_wine = drink['wine_servings'][mask].mean()
    print(f"{c}: beer={mean_beer:.2f}, wine={mean_wine:.2f}")

#5
print("\nTrung vị bia và rượu vang theo châu lục:")
for c in continents:
    mask = drink['continent']==c
    median_beer = np.median(drink['beer_servings'][mask])
    median_wine = np.median(drink['wine_servings'][mask])
    print(f"{c}: beer={median_beer}, wine={median_wine}")

#6
print("\nTiêu thụ rượu mạnh theo châu lục:")
for c in continents:
    spirit = drink['spirit_servings'][drink['continent']==c]
    print(f"{c}: mean={spirit.mean():.2f}, max={spirit.max()}, min={spirit.min()}")

#7
sorted_indices = np.argsort(drink['beer_servings'])
sorted_drink = drink[sorted_indices]

print("\n5 quốc gia uống bia ít nhất:")
print(sorted_drink[:5][['country','beer_servings']])

print("\n5 quốc gia uống bia nhiều nhất:")
print(sorted_drink[-5:][['country','beer_servings']])
```
