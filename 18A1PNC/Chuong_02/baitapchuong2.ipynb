```python
import requests
import xml.etree.ElementTree as ET
import csv

url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"
response = requests.get(url)

# Lưu RSS XML
with open("rss.xml", "wb") as f:
    f.write(response.content)

# Phân tích XML
tree = ET.parse("rss.xml")
root = tree.getroot()

# Mở file CSV
with open("rss_news.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Link", "Description"])

    for item in root.findall(".//item"):
        title = item.find("title").text
        link = item.find("link").text
        desc = item.find("description").text
        writer.writerow([title, link, desc])

print("Đã lưu tin tức vào rss_news.csv")
```

```python
import json

# Đọc file JSON
with open("nhanvien.json", "r", encoding="utf-8") as f:
    data = json.load(f)

company = data["company"]
name = company["name"]
address = company["address"]
departments = company["departments"]

# Tính tổng số nhân viên
total = sum(d["employees"] for d in departments)

print("Tên công ty:", name)
print("Địa chỉ:", address)
print("----- Thống kê nhân viên theo đơn vị -----")

for d in departments:
    percent = d["employees"] / total * 100
    print(f"Tên đơn vị: {d['name']}")
    print(f"  - Số nhân viên: {d['employees']}")
    print(f"  - Tỷ lệ: {percent:.2f}%\n")
```

```python
import json
from datetime import datetime

# Dữ liệu mẫu
giaodich = [
    {"loai": "vàng", "so_tien": 15000000},
    {"loai": "tiền mặt", "so_tien": 8000000}
]

chon = int(input("Có muốn lưu giao dịch không? (1=Có, 0=Không): "))

if chon == 1:
    now = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    filename = f"{now}.json"
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(giaodich, f, indent=4, ensure_ascii=False)
    print("Đã lưu giao dịch vào tệp:", filename)
else:
    print("Không lưu giao dịch.")
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
posts = response.json()

print("Tổng số bài post:", len(posts))
print("\nDanh sách các bài post:")

for post in posts[:5]:  # chỉ in 5 bài đầu
    print(f"userID: {post['userId']}, id: {post['id']}")
    print("Title:", post['title'])
    print("Body:", post['body'])
    print("-" * 40)
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/comments?postId=1"
response = requests.get(url)
comments = response.json()

print("Danh sách bình luận (3 cái đầu):")
for c in comments[:3]:
    print(f"ID: {c['id']}")
    print(f"Tên: {c['name']}")
    print(f"Email: {c['email']}")
    print(f"Nội dung: {c['body']}\n")
```
